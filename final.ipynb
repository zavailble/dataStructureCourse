# explanation of insert() and delMin()
### insert()
As for insert(), it's a method defined inside binaryHeap() class, take self and node two arguments.  
At first, it transform node into linked list node and append it into the linked list. append() is a method for add element in linked list.  
Then,it adopt size() method (a method for linked list size) to get new element index   
Next, it use swim() method to float the new element in the bottom of heap.   
```
def insert(self, node):
    self.append(node)
    index = self.size()
    self.swim(index)
```
In the swim() method, it apply getNode() to convert the index into a node. So, I can get the the node and its parent node.  
Then in order to have a smaller value in parent node, it compare node's value and its parent node's value.  
In while loop, if the node's value is less than its parent node's value, their value will be exchanged. Afterwards, 
Change the index of this node to the index of its parent node. The exchange function will be continuously excutived if it satisfies the compare comdition in while loop.  
```
    def swim(self, index: int):
        nodex = self.getNode(index)
        val = nodex.key
        parent = self.getNode(index // 2)
        if parent is not None:
            parentVal = parent.key
            while index > 0 and val < parentVal:
                temp = val
                nodex.key = parentVal
                parent.key = temp
                index = index // 2
                nodex = self.getNode(index)
                val = nodex.key
                parent = self.getNode(index // 2)
                if parent is not None:
                    parentVal = parent.key
                else:
                    break
```
